# 3. Imputasi

```python
#imputasi dengan mean
df3=df.fillna(df.mean())
df3.isna().sum()
```

```python
df3.describe()
```

```python
#imputasi dengan median
df4 = df.fillna(df.median())
df4.isna().sum()
```

```python
df4.describe()
```

```python
#imputasi dengan mode (untuk var kategorik)
df4['num-of-doors'] = df4['num-of-doors'].fillna(df4['num-of-doors'].mode()[0])
df4.isna().sum()
```
